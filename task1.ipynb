```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_76253.csv',skiprows=4)
df
```

```python
df.head()
```

```python
df.head(5)
```

```python
# select data for the year 2023
year = "2023"
df_2023 = df[["Country Name" , year]].dropna()
df_2023
```

```python
# sort the population and select top 10 countries for the bar chart
top_countries = df_2023.sort_values(by=year , ascending=False).head(10)
top_countries
```

```python
# create bar chart for top 10 most populated countries
plt.figure(figsize=(10,5))
plt.bar(top_countries["Country Name"],top_countries[year], color="skyblue" , edgecolor="black")
plt.xlabel("Country")
plt.ylabel("Population (in billions)")
plt.title(f"Top 10 Most Populated Countries in {year}")
plt.xticks(rotation=45)
plt.show()
           
```

```python
# Create histogram for population distribution
plt.figure(figsize=(8,5))
plt.hist(df_2023[year], bins=20 , edgecolor='m', alpha=0.7 , color="blue")
plt.xlabel("Population (in billions)")
plt.ylabel("Frequency")
plt.title(f"Population Distribution Across Countries in {year}")
plt.show()
```
